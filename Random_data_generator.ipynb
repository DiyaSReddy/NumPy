```python
import numpy as np
from datetime import datetime ,timedelta
```

```python
def sales_data_generator(no_samples = 100):
  np.random.seed(56) # fixed the random generated values
  product_ids = np.random.randint(1000,1130,no_samples)
  quantities = np.random.randint(1,200,no_samples)
  prices = np.random.uniform(500.4,999.5, no_samples)
  total_sales = (prices * quantities).round(2)
  starting_date = datetime(2025,1,1)
  dates = [(starting_date + timedelta(days=i)).strftime('%Y-%m-%d')for i in range (no_samples)]


  data  = list(zip(dates, product_ids , quantities , prices , total_sales))
  return data
```

```python
sales_data = sales_data_generator()
print("Sample data generated :")
for row in sales_data[:]:
  print(row)
```
