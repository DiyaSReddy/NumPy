```python
import numpy as np
```

```python
def create_mat():
  np.random.seed(36)
  data = np.random.randint(10, 101, 1000)
  print("Original Data: \n", data)
  return data
```

```python
def calc(data):
  print("\nMean:", np.mean(data))
  print("\nMedian:", np.median(data))
  print("\nStandard Deviation:", np.std(data))
  print("\nVariance:", np.var(data))
```

```python
def count_num(data):
  mean = np.mean(data)
  count = np.sum(data > mean)
  print("\nNumbers greater than the mean:", count)
```

```python
def normal(data):
  min = np.min(data)
  max = np.max(data)
  norm_data = (data - min) / (max - min)
  print("\nNormalized Data: \n", norm_data)
```

```python
def reshape(data):
  reshaped_data = data.reshape(20, 50)
  col_sums = np.sum(reshaped_data, axis=0)
  print("\nSum of each column: \n", col_sums)
```

```python
data = create_mat()
calc(data)
count_num(data)
normal(data)
reshape(data)
```
